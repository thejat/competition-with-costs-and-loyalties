```python
#keep cb fixed and ca is varied
flag_plot = True
flag_theory = True
dist = 'uniform'
cb = 0.4
la = 1
lb = 4
sa = 3
sb = 3
theta = 0.8
npts = 40 
ca_arr = np.linspace(cb,10,100) #ca >= cb, hardcode
maxpx=ca_arr.max()+5
df = ll_get_metric_arrs_vs_camcb(ca_arr,cb,la,lb,sa,sb,maxpx,npts,dist,theta,flag_theory=True)
ca_m_cb_arr = ca_arr - cb
```

```python
vert_line_locs1 = [sa-la,sa+2*la]
vert_line_locs1_labels = [r'$c_a - c_b = s_{\alpha}-l_{\alpha}$',r'$c_a-c_b=s_a+2l_{\alpha}$']
labels1=[r'$p_{A}^{\alpha}$',r'$p_{B}^{\alpha}$']
fname_prices_a = "./ll_single_stage_paa_pba.png" if flag_plot else None
plot_curves_vs_camcb(df['paa'],df['pba'],ca_m_cb_arr,'Prices',labels1,vert_line_locs1,vert_line_locs1_labels,fname_prices_a)
```

```python
vert_line_locs2 = [lb-sb]
labels2=[r'$p_{B}^{\beta}$',r'$p_{A}^{\beta}$']
vert_line_locs2_labels = [r'$c_a - c_b = l_{\beta}-s_{\beta}$']
fname_prices_b = "./ll_single_stage_pbb_pab.png" if flag_plot else None
plot_curves_vs_camcb(df['pbb'],df['pab'],ca_m_cb_arr,'Prices',labels2,vert_line_locs2,vert_line_locs2_labels,fname_prices_b)
```

```python
vert_line_locs3 = [lb-sb,sa-la,sa+2*la]
labels3=[r'Firm A',r'Firm B']
fname_market_share = "./ll_single_stage_market_share.png" if flag_plot else None
plot_curves_vs_camcb(df['marketshare_a'],df['marketshare_b'],ca_m_cb_arr,'Market Share',labels3,vert_line_locs3,None,fname_market_share)
```

```python
vert_line_locs3 = [lb-sb,sa-la,sa+2*la]
labels3=[r'Firm A',r'Firm B']
fname_profits = "./ll_single_stage_profits.png" if flag_plot else None
plot_curves_vs_camcb(df['total_profit_a'],
                     df['total_profit_b'],ca_m_cb_arr,'Profit',labels3,vert_line_locs3,None,fname_profits)
```
